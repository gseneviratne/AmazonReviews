# <font color = purple> Apache Spark <font color = black>
## (Almost) All in one

Spark, an open-source, distributed computing system, plays a fundamental role in this project's data processing phase. It enables data enrichment by incorporating the following key attributes:
    
* Lyrics with Genius API
* Polarity - Subjectivity with TextBlob
* Prediction with KMeans clustering

# <font color = purple> Sentiment Analysis and Clustering <font color = black>

## Discovering the power of grouping

For the sentiment analysis, has been involved a pre-trained model loaded from the *SpaCy* library.
Then has been written two functions that utilize the *TextBlob* library to calculate the polarity and subjectivity of a given set of lyrics.

For the clustering phase, the implementation proposed, utilizes the *KMeans* algorithm from the *Spark ML library*.
